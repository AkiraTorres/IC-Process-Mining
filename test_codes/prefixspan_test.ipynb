```python
with open('./sceneries/1-first.json') as file:
    data = json.load(file)
```

```python
# Extract sequences of events
sequences = []
for entry in data:
    events = [event['event'] for event in entry['events']]
    sequences.append(events)
```

```python
ps = PrefixSpan(sequences)
min_support = len(sequences)*0.08
```

```python
frequent_sequences = ps.frequent(min_support)
```

```python
for sequence, support in frequent_sequences:
    print(f"Sequence: {sequence}, Support: {support}")
```
